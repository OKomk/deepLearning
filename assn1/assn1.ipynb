```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        # Initialize weights and bias to zero
        self.weights = np.zeros(X.shape[1])
        self.bias = 0
        
        # Gradient descent algorithm
        for i in range(self.num_iterations):
            y_pred = self.predict(X)
            insterror = 0.5*((y-y_pred)**2)
            error = y - y_pred
            self.weights += self.learning_rate * np.dot(X.T, error) # updating the weights
            self.bias += self.learning_rate * np.sum(error)
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
# Generate a sample dataset
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.2 * np.random.randn(100, 1)

# Fit a linear regression using a single perceptron
model = Perceptron(learning_rate=0.01, num_iterations=1000)
model.fit(X, y)

# Make predictions on new data
X_test = np.array([[0.1], [0.5], [0.9]])
# y_pred = model.predict(X_test)
print (X_test.shape[1])
print (X_test.shape)
# Print the learned weights and predictions
# print("Weights:", model.weights)
# print("Bias:", model.bias)
# print("Predictions:", y_pred)
```
